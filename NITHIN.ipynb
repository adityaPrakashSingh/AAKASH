```python
#list of avg rides of different bases in 6 consecutive months
a=[0]*6
a[0]=(uber_april.groupby(['Base']).size()//30).tolist()
a[1]=(uber_may.groupby(['Base']).size()//30).tolist()
a[2]=(uber_jun.groupby(['Base']).size()//30).tolist()
a[3]=(uber_jul.groupby(['Base']).size()//30).tolist()
a[4]=(uber_aug.groupby(['Base']).size()//30).tolist()
a[5]=(uber_sep.groupby(['Base']).size()//30).tolist()
a
```

```python
#bar graph avg rides of different bases in 6 consecutive months
X = np.arange(5)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(X + 0.00, a[0], color = 'b', width = 0.12)
ax.bar(X + 0.12, a[1], color = 'g', width = 0.12)
ax.bar(X + 0.24, a[2], color = 'r', width = 0.12)
ax.bar(X + 0.36, a[3], color = 'yellow', width = 0.12)
ax.bar(X + 0.48, a[4], color = 'violet', width = 0.12)
ax.bar(X + 0.60, a[5], color = 'orange', width = 0.12)
ax.set_ylabel('avg rides')
ax.set_xlabel(' B02512            B02598            B02617            B02682            B02764 ')
ax.set_title('Avg rides per day of different cabs in particular months')
ax.legend(labels=['apr','may','jun','jul','aug','sep'])
#ax.set_xticks(X,('B02512','B02598','B02617','B02682','B02764'))
```
